下記のプロンプトの内容を読み上げます

```python
#読み上げるプロンプトを設定する。

prompt = """
こんにちは！
はじめまして。
"""
```

```python

from IPython.display import display, Audio
sbv2 = SBV2()
sr, audio = sbv2.call_TTS(prompt)
display(Audio(audio, rate=sr))
```

```python
import soundfile
soundfile.write(f"./outputs/SBV2_output.wav",
                audio, sr, "PCM_16")
```
